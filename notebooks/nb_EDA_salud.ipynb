## salud_mental

```python
#función para categorizar salud mental 
def salud_mental(x):
    if x == 0:
        return 'no depresion'
    elif x >= 1 and x <= 4:
        return 'depresion minima'
    elif x >= 5 and x <= 9:
        return 'depresion leve'
    elif x >= 10 and x <= 14:
        return 'depresion moderada'
    elif x >= 15 and x <= 19:
        return 'depresion moderadamente severa'
    elif x >= 20 and x <= 27:
        return 'depresion severa'
    else:
        return 'error en suma'
```

```python
#función para sumar por fila las respuestas de poco_interes_14d, deprimido_14d, problema_dormir_14d, cansancio_14d, problema_comer_14d, problema_atencion_14d, sentirse_mal_14d y pensamiento_suicida_14d
salud['phq9'] = salud.iloc[:, 7:16].sum(axis=1)
```

```python
#aplicar función a la columna phq9
salud['salud_mental'] = salud['phq9'].apply(salud_mental)
```

```python
# eliminar las columnas 'poco_interes_14d', 'deprimido_14d', 'problema_dormir_14d','cansancio_14d', 'problema_comer_14d', 'problema_atencion_14d','problema_moverse_14d', 'pensamiento_suicida_14d', 'sentirse_mal_14d' y phq9
salud = salud.drop(['poco_interes_14d', 'deprimido_14d', 'problema_dormir_14d','cansancio_14d', 'problema_comer_14d', 'problema_atencion_14d','problema_moverse_14d', 'pensamiento_suicida_14d', 'sentirse_mal_14d', 'phq9'], axis=1)
salud
```
