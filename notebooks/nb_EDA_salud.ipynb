# Columnas

## salud_mental

```python
#función para categorizar salud mental 
def salud_mental(x):
    if x == 0:
        return 'no depresion'
    elif x >= 1 and x <= 4:
        return 'depresion minima'
    elif x >= 5 and x <= 9:
        return 'depresion leve'
    elif x >= 10 and x <= 14:
        return 'depresion moderada'
    elif x >= 15 and x <= 19:
        return 'depresion moderadamente severa'
    elif x >= 20 and x <= 27:
        return 'depresion severa'
    else:
        return 'error en suma'
```

```python
#Sumar por fila las respuestas de la escala PHQ-9
salud['phq9'] = salud.iloc[:, 7:16].sum(axis=1)
```

```python
#aplicar función de salud mental
salud['salud_mental'] = salud['phq9'].apply(salud_mental)
```

```python
# eliminar las columnas de las preguntas de la escala PHQ-9
salud = salud.drop(['poco_interes_14d', 'deprimido_14d', 'problema_dormir_14d','cansancio_14d', 'problema_comer_14d', 'problema_atencion_14d','problema_moverse_14d', 'pensamiento_suicida_14d', 'sentirse_mal_14d', 'phq9'], axis=1)
salud
```

## categoria_imc

```python
#función para categorizar imc
def obesidad(x):
    if x < 18.5:
        return 'bajo'
    elif x >= 18.5 and x <= 24.9:
        return 'normal'
    elif x >= 25 and x <= 29.9:
        return 'sobrepeso'
    elif x >= 30 and x <= 34.9:
        return 'obesidad 1'
    elif x >= 35 and x <= 39.9:
        return 'obesidad 2'
    else:
        return 'obesidad 3'
```

```python
#cálculo de imc
salud["imc"] = salud["peso_kg"] / (salud["talla_cm"] / 100) ** 2
salud["imc"] = salud["imc"].round(2)
```

```python
#aplicar función de imc
salud['categoria_imc'] = salud['imc'].apply(obesidad)
salud
```

## diagnostico_obesidad

```python
#función para diagnostico obesidad 
def diagnostico_obesidad(x):
    if x >= 30:
        return 'si'
    else:
        return 'no'
```

```python
#aplicar función de diagnostico obesidad
salud['diagnostico_obesidad'] = salud['imc'].apply(diagnostico_obesidad)
```

```python
#elminar columnas de peso, talla e imc
salud = salud.drop(['peso_kg', 'talla_cm', 'imc'], axis=1)
salud
```

## categoria_presion

```python
#función para categoria presion
def categoria_presion (sistolica,diastolica):
    if sistolica < 120 and diastolica < 80:
        return "normal"
    elif sistolica >= 120 and sistolica <=129  and diastolica < 80:
        return "elevado"
    elif sistolica >= 130 and sistolica <=139  or diastolica >= 80 and diastolica <= 89:
        return "hipertension etapa 1"
    elif sistolica >= 140 and sistolica < 180 or diastolica >= 90 and diastolica <= 120:
        return "hipertension etapa 2"
    elif sistolica >= 180 or diastolica > 120:
        return "crisis hipertensiva"
    else:
        return "error"
```

```python
#calcular el promedio de la presion sistolica y diastolica
salud["media_sis"] = salud[["presion_sis_1", "presion_sis_2"]].mean(axis=1)
salud["media_dia"] = salud[["presion_dia_1", "presion_dia_2"]].mean(axis=1)

#redondear media_sis y media dia a enteros
salud["media_sis"] = salud["media_sis"].round(0).astype(int)
salud["media_dia"] = salud["media_dia"].round(0).astype(int)
```

```python
#aplicar función de categoria presion
salud['categoria_presion'] = salud.apply(lambda x: categoria_presion(x['media_sis'], x['media_dia']), axis=1)
```

```python
#eliminar columnas de presiones y medias de presiones
salud = salud.drop(['presion_sis_1', 'presion_sis_2', 'presion_dia_1', 'presion_dia_2', 'media_sis', 'media_dia'], axis=1)
salud
```
